```python
# Create list of numerical columns df for outlier detection
numerical_columns = train.select_dtypes(include=['number']).columns.tolist()

time_columns = ['YearBuilt', 'YearRemodAdd', 'GarageYrBlt', 'YrSold', 'MoSold']
numerical_columns = [col for col in numerical_columns if col not in time_columns]

categorical_columns = ['MSSubClass', 'OverallQual', 'OverallCond', 'HasPool', 'HasFireplace', 'HasFence']
numerical_columns = [col for col in numerical_columns if col not in categorical_columns]

target_column = ['SalePrice']
numerical_columns = [col for col in numerical_columns if col not in target_column]

numerical_columns
```

```python
# Define function that removes outliers based on the IQR method
def remove_outliers(df, numerical_columns):
    for numerical_column in numerical_columns:
        Q1 = df[numerical_column].quantile(0.25)
        Q3 = df[numerical_column].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Remove rows where the numerical_column's values are outside the bounds
        df = df[(df[numerical_column] >= lower_bound) & (df[numerical_column] <= upper_bound)]

    return df
```

```python


# Define helper functions
def replace_mode(df, column):
    """
    Replace missing values in a column with the mode of the column.
    """
    mode_value = df[column].mode()[0]  # Calculate the mode
    df[column] = df[column].fillna(mode_value)

def preprocess_data(df):
    label_encoder = LabelEncoder()

    # Handle missing values for categorical columns before encoding
    categorical_cols = df.select_dtypes(include=['object']).columns  # Identify categorical columns
    for col in categorical_cols:
        if df[col].isnull().sum() > 0:
            replace_mode(df, col)  # Replace missing values with mode

        # Apply Label Encoding
        df[col] = label_encoder.fit_transform(df[col])

    # Handle missing values for numerical columns
    for col in df.columns:
        if df[col].isnull().sum() > 0:
            replace_mode(df, col)

    # Normalize numerical features using MinMaxScaler
    numeric_cols = df.select_dtypes(include=[np.number]).columns  # Identify numerical columns
    scaler = MinMaxScaler()
    df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

    return df

# Load train and test datasets
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

# Keep 'Id' column in the test dataset for final submission
test_ids = test['Id']

# Apply preprocessing
train = preprocess_data(train)
test = preprocess_data(test)

# Ensure test data has the same features as training data
common_features = train.drop(columns=['SalePrice']).columns.intersection(test.columns).tolist()

# Align features and drop target for train
X_train = train[common_features]
y_train = train['SalePrice']

# Align features for test
X_test = test[common_features]

# Split training data for validation
X_train, X_val, y_train, y_val = train_test_split(
    X_train, y_train, test_size=0.2, random_state=42
)

# Define models
models = {
    "KNN Regressor": KNeighborsRegressor(),
    "Random Forest": RandomForestRegressor(),
    "XGBoost": XGBRegressor()
}

# Model comparison list
model_comparison = []

def evaluate_model_and_store(model, model_name, X_train, y_train, X_val, y_val):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_val)

    # Calculate metrics
    mae = mean_absolute_error(y_val, y_pred)
    mse = mean_squared_error(y_val, y_pred)
    r2_train = r2_score(y_train, model.predict(X_train))
    r2_val = r2_score(y_val, y_pred)

    # Append results
    model_comparison.append({
        'Model': model_name,
        'MAE': mae,
        'MSE': mse,
        'R2_Score (Train)': r2_train,
        'R2_Score (Val)': r2_val
    })

# Evaluate each model
for name, model in models.items():
    print(f"\nTraining and evaluating {name}...")
    evaluate_model_and_store(model, name, X_train, y_train, X_val, y_val)

# Create DataFrame for model comparison
compare_models = pd.DataFrame(model_comparison)

# Display comparison results
print(compare_models)

# Select best model (example: XGBoost)
best_model = XGBRegressor(subsample=0.8, n_estimators=100, max_depth=7, learning_rate=0.05, colsample_bytree=0.6)
best_model.fit(X_train, y_train)

# Predict on the test dataset
test_predictions = best_model.predict(X_test)

# Prepare submission DataFrame
test_results = pd.DataFrame({
    'Id': test_ids,  # Use stored 'Id' from test set
    'SalePrice': test_predictions
})

# Save predictions to CSV
test_results.to_csv('house_price_predictions.csv', index=False)
print("Predictions saved to house_price_predictions.csv")
```
